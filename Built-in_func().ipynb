```python
#lambda: are a key concept in python and extremely useful for short,
# throwaway functions - especially in sorting, filtering, and mapping data.

# lambda arguements: expression
# lambda - keyword
# arguements - input (like parameters)
# expression - single expression whose result is returned automatically.

def sqr(x): 
    return x**2
sqr(4)

# Using Lambda function
sqr1 = lambda x:x**2
print("Square: ",sqr1(2))

# Multiple Arguements:
add = lambda a,b : a+b 
print("Sum: ",add (3,5),type(add))
```

```python
# Where lambda's are commonly used 
# 1. in sorted() or sort()
# 2. with map()
# 3. with filter()

#no return value
# type gives function
# one line code
# not used for code reusability
# no name


###Limitations: 
# 1. only one expression (no arguements,loops, or multiple line)
# 2. Often best used for simple operations
# 3. For complex logic: use a regular def function

l = ['apple','banana','cherry']
l1 = []
for i in l: 
    l1.append(i[0])

print(l1)
l = 'apple'
b = lambda l:l[0]
print(b(l))

x = 'apple'
def log (x):
    return x[1] == 'a'
log('apple')

b = lambda x : x[0] == 'a'
print(b(x))
```

```python
#decorator: is a function that take another function as arguement 
# and return a function
# built in decorator: Already defined
# user defined generator: generator are user defined

# Decorator: takes functions as a arguements 
# and return function also

def func(): 
    print("hello")
x=func
x()

```

```python

# Decorate a gift item with a packing sheet same goes to decorator 
# means wrapping a function using another function which is taken as arguements
# for adding some new information also

def decorator (func): # return none
    def wrapper():
        print("Start of function")
        func()
        print("End of transaction")
        # return none
    return wrapper()

def sum():
    print("hello") # return none
    
decorator(sum)
```

```python
# Decorate a gift item with a packing sheet same goes to decorator 
# means wrapping a function using another function which is taken as arguements
# for adding some new information also

from functools import wraps

def decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Start of function")
        result = func(*args, **kwargs)
        print("End of transaction")
        return result
    return wrapper

def sum1(a):
    """Adds 10 to a and says hello."""
    print("hello", a)
    return a 
    
@decorator
sum1 = decorator(sum1)
# # call normally
# print("Function name:", sum1.__name__)
# print("Docstring   :", sum1.__doc__)
print("Call result :", sum1(4))

```

```python
def decorator(func): 
    def wrapper(x,y):
        print("Inside the Decorator")
        r= func(x,y)
        print(r)
        print("End of the sum operation")
        return r
    return wrapper

def sum2(a,b):
    print("Inside sum()")
    return a+b

sum3 = decorator(sum2)
print(sum2(2,4))

```
